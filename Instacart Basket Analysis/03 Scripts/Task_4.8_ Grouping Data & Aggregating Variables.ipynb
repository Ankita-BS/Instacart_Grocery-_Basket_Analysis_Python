### 6.

The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
# Creating a new column mean_price
ords_prods_merge['mean_price'] = ords_prods_merge.groupby('user_id')['prices'].transform(ny.mean)
```

```python
ords_prods_merge.head()
```

```python
#Creating new column spender_type based on average expenditure
ords_prods_merge.loc[ords_prods_merge['mean_price']< 10 , 'spender_type'] = 'Low Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price']>= 10 , 'spender_type'] = 'High Spender'
```

```python
#Checking frequency
ords_prods_merge['spender_type'].value_counts(dropna = False)
```

```python
#Checking relavent columns
ords_prods_merge[['user_id','mean_price','spender_type']].head(25)
```

### 7. Create frequency flag to determine frequent and non frequent flag

In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
#Creating new column median_prior_order
ords_prods_merge['median_prior_order'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform(ny.median)
```

```python
ords_prods_merge.head()
```

```python
# Create order frequency flag based on given criteria using loc function
ords_prods_merge.loc[ords_prods_merge['median_prior_order']> 20 , 'frequency'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_prior_order']> 10) & (ords_prods_merge['median_prior_order']<= 20) , 'frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_prior_order'] <= 20 , 'frequency'] = 'Frequent customer'
```

```python
#Checking freqency of the new column
ords_prods_merge['frequency'].value_counts(dropna = False)
```

There are 5 values for NAN , which may be due to bank values in 'days since prior order'. As it is first order the values are blank. 

```python
# Checking relevant columns
ords_prods_merge[['user_id','median_prior_order','frequency']].head(25)
```
